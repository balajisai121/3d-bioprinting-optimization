```python
# Python Script: Enhanced Bioprinting Data Analysis and Model Pipeline
# This script visualizes data, identifies correlations, and develops models
# (SVM, Neural Network, and additional models) to predict and optimize bioprinting quality.

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelBinarizer
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, ReLU
from tensorflow.keras.optimizers import Adam

# Step 1: Load and Inspect Data
# Load the cleaned data file
cleaned_data = pd.read_csv('cleaned.csv')

# Display the first few rows of the dataset
print('Preview of the Data:')
print(cleaned_data.head())

# Step 2: Data Visualization
# Visualizing pairwise correlations among numerical variables
plt.figure(figsize=(12, 10))
sns.heatmap(cleaned_data.corr(), annot=True, fmt='.2f', cmap='coolwarm', cbar=True)
plt.title('Correlation Heatmap')
plt.xlabel('Features')
plt.ylabel('Features')
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.tight_layout()
plt.show()

# Histogram for Print Quality distribution
plt.figure()
sns.histplot(cleaned_data['Print Quality'], kde=True, color='blue')
plt.title('Distribution of Print Quality')
plt.xlabel('Print Quality')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

# Step 3: Define Target and Features (Updated for Classification)
# Define features (e.g., material properties, print parameters)
features = cleaned_data.drop(columns=['Print Quality'])
target = cleaned_data['Print Quality']

# Classify Print Quality into 'Good', 'Bad', and 'Best'
# Define thresholds for classification (e.g., based on quantiles)
q1 = target.quantile(0.33)  # Lower third
q2 = target.quantile(0.66)  # Upper third

print_quality_class = pd.cut(target, bins=[-np.inf, q1, q2, np.inf], labels=['Bad', 'Good', 'Best'])

# Update the target variable for classification
target = print_quality_class

# Step 4: Preprocessing for Models
# Standardize numeric features for consistent scaling
scaler = StandardScaler()
standardized_features = scaler.fit_transform(features)

# Step 5: Train-Test Split
x_train, x_test, y_train, y_test = train_test_split(standardized_features, target, test_size=0.3, random_state=42)

# Step 6: Train and Evaluate Models
models = {
    'SVM': SVC(kernel='rbf', probability=True),
    'Logistic Regression': LogisticRegression(),
    'Random Forest': RandomForestClassifier(),
    'Gradient Boosting': GradientBoostingClassifier(),
    'Decision Tree': DecisionTreeClassifier(),
    'KNN': KNeighborsClassifier()
}

results = {}
for name, model in models.items():
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    accuracy = accuracy_score(y_test, y_pred)
    results[name] = accuracy
    print(f'{name} Accuracy: {accuracy * 100:.2f}%')
    print(f'Classification Report for {name}:\n', classification_report(y_test, y_pred))

# Step 7: Neural Network Model
# Define Neural Network architecture
nn_model = Sequential([
    Dense(10, input_dim=x_train.shape[1]),
    ReLU(),
    Dense(3, activation='softmax')  # Output layer with 3 units (for 3 classes) and softmax activation
])

# Compile the model
nn_model.compile(optimizer=Adam(learning_rate=0.001), loss='categorical_crossentropy', metrics=['accuracy'])

# Convert target labels to one-hot encoding
encoder = LabelBinarizer()
y_train_encoded = encoder.fit_transform(y_train)
y_test_encoded = encoder.transform(y_test)

# Train Neural Network
nn_model.fit(x_train, y_train_encoded, epochs=50, batch_size=32, verbose=1)

# Evaluate Neural Network
predicted_nn = nn_model.predict(x_test)
nn_accuracy = accuracy_score(np.argmax(y_test_encoded, axis=1), np.argmax(predicted_nn, axis=1))
print(f'Neural Network Accuracy: {nn_accuracy * 100:.2f}%')

# Step 8: Optimization Insights
# Feature importance visualization
feature_importance = pd.DataFrame({'Feature': features.columns, 'Importance': np.mean(np.abs(scaler.inverse_transform(x_train)), axis=0)})
feature_importance = feature_importance.sort_values(by='Importance', ascending=False)

plt.figure(figsize=(12, 8))
sns.barplot(data=feature_importance, x='Importance', y='Feature', palette='viridis')
plt.title('Feature Importance for Print Quality')
plt.xlabel('Importance')
plt.ylabel('Features')
plt.tight_layout()
plt.show()

# Step 9: Visualize Predicted vs Actual Results for Neural Network
plt.figure()
plt.scatter(np.argmax(y_test_encoded, axis=1), np.argmax(predicted_nn, axis=1), color='blue', label='Predictions')
plt.plot([0, 2], [0, 2], 'r--', label='Ideal')
plt.title('Neural Network Predictions vs Actual')
plt.xlabel('Actual Print Quality Class')
plt.ylabel('Predicted Print Quality Class')
plt.legend()
plt.tight_layout()
plt.show()

# Print Model Results
print('Model Results:')
for model_name, accuracy in results.items():
    print(f'{model_name}: {accuracy * 100:.2f}%')
```
